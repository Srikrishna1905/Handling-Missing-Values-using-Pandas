# Filling Null ***values***

```python
#Filling Null values with 0 value
df2 = df.fillna(value = 0 )
df2
```

```python
#Filling Null values with 1 value
df3 = df.fillna(value = 1)
df3
```

#*Filling* Null values with a previous value in row wise

```python
df4 = df.fillna(method = 'pad')
df4
```

# Filling NUll values with the next value in row wise

```python
df5 = df.fillna(method = 'bfill')
df5
```

# Filling Null values with the previous value in column wise

```python
df6 = df.fillna(method = 'pad', axis = 1)
df6
```

# Filling Null values with the next value in column wise

```python
df7 = df.fillna(method = 'bfill', axis = 1)
df7
```

# Filling Different values in null in different column

```python
df8 = df.fillna({'society':'abcd', 'balcony':'xyz'})
df8
```

# Filling Null value with the mean of a column

```python
df9 = df.fillna(value = df['balcony'].min()) # we can also use max, mean in the position of min and we will get different outputs
df9
```

## dropna()


```python
df
```

Delete all rows which have Null value.

```python
df10 = df.dropna()
df10
```

Delete all rows which have at least 1 Null value.

```python
df11 = df.dropna(how = 'any')
df11
```

Delete all rows in which all values are Null.

```python
df12 = df.dropna(how = 'all')
df12
```

#replace()

```python
import numpy as np
df12 = df.replace(to_replace = np.nan, value = "Str")
df12
```

```python
df13 = df.replace(to_replace = 3.0, value = 814576 )
df13
```
